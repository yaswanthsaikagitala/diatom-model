```python
import numpy as np
import joblib
from imblearn.over_sampling import SMOTE, ADASYN
```

```python
def balance_data(X_train, y_train):
    print("Balancing Data...")
    print(f"Class distribution before balancing: {np.bincount(y_train)}")
    try:
        adasyn = ADASYN(sampling_strategy='not majority', random_state=42)
        X_train_res, y_train_res = adasyn.fit_resample(X_train, y_train)
        print("✅ ADASYN used for balancing.")
    except ValueError:
        smote = SMOTE(sampling_strategy='not majority', random_state=42)
        X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
        print("✅ SMOTE used for balancing.")
    print(f"Class distribution after balancing: {np.bincount(y_train_res)}")
    print(f"X_train_res shape: {X_train_res.shape}, y_train_res shape: {y_train_res.shape}")
    return X_train_res, y_train_res
```

```python
# Load preprocessed data
X_train, y_train = joblib.load("X_train_preprocessed.pkl")
X_test, y_test   = joblib.load("X_test_preprocessed.pkl")
X_val, y_val     = joblib.load("X_val_preprocessed.pkl")
class_names      = joblib.load("class_names.pkl")

# Balance training data
X_train_res, y_train_res = balance_data(X_train, y_train)
```

```python
# Save balanced data into .pkl files
joblib.dump((X_train_res, y_train_res), "X_train_balanced.pkl")
joblib.dump((X_test, y_test), "X_test_balanced.pkl")
joblib.dump((X_val, y_val), "X_val_balanced.pkl")
joblib.dump(class_names, "class_names.pkl")

print("✅ Balanced data saved as .pkl files")
print("X_train_res:", X_train_res.shape, "y_train_res:", y_train_res.shape)
print("X_test:", X_test.shape, "y_test:", y_test.shape)
print("Class Names:", class_names)
```

- Prints class distribution before and after balancing.
- Confirms whether ADASYN or SMOTE was used.
- Saves balanced training set (X_train_balanced.pkl), test set (X_test_balanced.pkl), and class names (class_names.pkl).
- Prints dataset shapes for verification
